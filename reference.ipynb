```python
from gensim.models import KeyedVectors
model_word = KeyedVectors.load_word2vec_format('id.vec')
```

```python
key = 'jahat'
similar = model_word.similar_by_word(key, 20)
print(similar)
```

```python
outcome = []
count = 0
number = 1
for i, j in similar:
    if count<5:
        for line in open('quran.txt'):
            if i in line:
                outcome.append(line.strip())

        if not outcome:
            print(str(number) +' not')
        else:
            count+=1
            print(str(number)+ ' found ' + i + ' on order ' + str(count))
            print(outcome)
            print('\n')
            
        number+=1

        outcome.clear()
```

```python
key = 'jahat'
similar = model_word.similar_by_word(key, 20)
n = 1
for i, j in similar:
    print(str(n)+ ' ' + i)
    n+=1
```

```python
response = []
outcome = []
count = 0
for i, j in similar:
    if count<10:
        for line in open('quran.txt'):
            if i in line:
                outcome.append(line.strip())

        if outcome:
            count+=1
            response.append(i)

        outcome.clear()

print(response)
```
